```python
import pandas as pd
```

```python
pp1 = pd.read_csv('data/pp/pp-2014-part1.csv', header=None)
pp2 = pd.read_csv('data/pp/pp-2014-part2.csv', header=None)
```

```python
# take 10 first chars from column number 2 and convert to datetime
pp1['Date'] = pd.to_datetime(pp1[2].str[:10], format='%Y-%m-%d')
pp2['Date'] = pd.to_datetime(pp2[2].str[:10], format='%Y-%m-%d')
```

```python
pd.concat([pp1[pp1['Date'] < '2014-07-01'], pp2[pp2['Date'] < '2014-07-01']]).to_csv('data/pp/prepared/pp-2014-part1.csv', index=False, header=False)
pd.concat([pp1[pp1['Date'] >= '2014-07-01'], pp2[pp2['Date'] >= '2014-07-01']]).to_csv('data/pp/prepared/pp-2014-part2.csv', index=False, header=False)
```

```python
years = ['2014', '2015', '2016', '2017']

for year in years:
    pp1 = pd.read_csv(f'data/pp/pp-{year}-part1.csv', header=None)
    pp2 = pd.read_csv(f'data/pp/pp-{year}-part2.csv', header=None)
    pp1['Date'] = pd.to_datetime(pp1[2].str[:10], format='%Y-%m-%d')
    pp2['Date'] = pd.to_datetime(pp2[2].str[:10], format='%Y-%m-%d')
    pd.concat([pp1[pp1['Date'] < f'{year}-04-01'], pp2[pp2['Date'] < f'{year}-04-01']]).drop(['Date'], axis=1).to_csv(f'data/pp/prepared/pp-{year}-q1.csv', index=False, header=False)
    pd.concat([pp1[(pp1['Date'] >= f'{year}-04-01') & (pp1['Date'] < f'{year}-07-01')], pp2[(pp2['Date'] >= f'{year}-04-01') & (pp2['Date'] < f'{year}-07-01')]]).drop(['Date'], axis=1).to_csv(f'data/pp/prepared/pp-{year}-q2.csv', index=False, header=False)
    pd.concat([pp1[(pp1['Date'] >= f'{year}-07-01') & (pp1['Date'] < f'{year}-10-01')], pp2[(pp2['Date'] >= f'{year}-07-01') & (pp2['Date'] < f'{year}-10-01')]]).drop(['Date'], axis=1).to_csv(f'data/pp/prepared/pp-{year}-q3.csv', index=False, header=False)
    pd.concat([pp1[pp1['Date'] >= f'{year}-10-01'], pp2[pp2['Date'] >= f'{year}-10-01']]).drop(['Date'], axis=1).to_csv(f'data/pp/prepared/pp-{year}-q4.csv', index=False, header=False)
```

```python
year = '2018'
pp = pd.read_csv(f'data/pp/pp-{year}.csv', header=None)
pp['Date'] = pd.to_datetime(pp[2].str[:10], format='%Y-%m-%d')
pp[pp['Date'] < f'{year}-04-01'].drop(['Date'], axis=1).to_csv(f'data/pp/prepared/pp-{year}-q1.csv', index=False, header=False)
pp[(pp['Date'] >= f'{year}-04-01') & (pp['Date'] < f'{year}-07-01')].drop(['Date'], axis=1).to_csv(f'data/pp/prepared/pp-{year}-q2.csv', index=False, header=False)
pp[(pp['Date'] >= f'{year}-07-01') & (pp['Date'] < f'{year}-10-01')].drop(['Date'], axis=1).to_csv(f'data/pp/prepared/pp-{year}-q3.csv', index=False, header=False)
pp[pp['Date'] >= f'{year}-10-01'].drop(['Date'], axis=1).to_csv(f'data/pp/prepared/pp-{year}-q4.csv', index=False, header=False)
```
